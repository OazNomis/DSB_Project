```python
df['issue_d'] = pd.to_datetime(df['issue_d'])
df = df[df['issue_d']>='2016-01-01']
```

```python
df.describe()
```

```python
#Removing string from term
df['term'] = df['term'].astype(str).str.replace(' months', '')
df['term'] = df['term'].apply(lambda x:int(x) if x != 'NaN')
df['emp_length'] = df['emp_length'].astype(str).str.replace('+ years','')

#cleaning to last three years only
df = df[df['issue_d'] >= '2016-01-01']
```

```python
df['grade']= pd.get_dummies(df['grade'],drop_first=True)
df['sub_grade'] = pd.get_dummies(df['sub_grade'],drop_first=True)
#df['emp_title'] = pd.get_dummies(df['emp_title'],drop_first=True)
df['home_ownership'] = pd.get_dummies(df['home_ownership'],drop_first=True)
df['verification_status'] = pd.get_dummies(df['verification_status'],drop_first=True)
df['issue_d'] = pd.to_datetime(df['issue_d'])
df['loan_status'] = pd.get_dummies(df['loan_status'],drop_first=True)
df['pymnt_plan'] = pd.get_dummies(df['pymnt_plan'],drop_first=True)
df['purpose'] = pd.get_dummies(df['purpose'],drop_first=True)
#df['title'] = pd.get_dummies(df['title'],drop_first=True)
# not sure what to do about zip_code
df['addr_state'] = pd.get_dummies(df['addr_state'],drop_first=True)
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'])
df['initial_list_status'] = pd.get_dummies(df['initial_list_status'],drop_first=True)
df['last_pymnt_d'] = pd.get_dummies(df['last_pymnt_d'],drop_first=True)
df['next_pymnt_d'] = pd.get_dummies(df['next_pymnt_d'], drop_first=True)
df['last_credit_pull_d'] = pd.to_datetime(df['last_credit_pull_d'])
df['application_type'] = pd.get_dummies(df['application_type'],drop_first=True)
df['hardship_flag'] = pd.get_dummies(df['hardship_flag'],drop_first=True)
df['disbursement_method'] = pd.get_dummies(df['disbursement_method'], drop_first=True)
df['debt_settlement_flag'] = pd.get_dummies(df['debt_settlement_flag'],drop_first=True)
```
